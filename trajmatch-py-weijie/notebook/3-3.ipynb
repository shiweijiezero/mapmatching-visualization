```python
%matplotlib inline

import pandas as pd

DATA = []

for type_ in ['-HMM','-CRF','-INT']:
    fname = f'3-3/3-3.txt{type_}'
    data = pd.read_csv(fname, sep='\t', header=None)

    data = data[(data[1] > -1) & (data[1] < 3)]
    print(fname)
    DATA.append(data[1])
#     print(data.describe())
#     data[1].hist(bins=100,figsize=(12,8),alpha=0.5)
hmm_label = f"HMM-{DATA[0].mean()}"
crf_label = f"CRF-{DATA[1].mean()}"
int_label = f"INI-{DATA[2].mean()}"
d = pd.DataFrame({hmm_label: DATA[0],crf_label:DATA[1],int_label:DATA[2]}, columns=[hmm_label,crf_label,int_label])
d.plot.hist(alpha=0.6,bins=200,figsize=(12,8))
    
    
```
